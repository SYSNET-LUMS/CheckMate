```python
import os
```

```python
path_to_input = "input_images/"
path_to_output = "output_images/"
```

```python
truncation_values = [0, 5, 10, 15, 20, 25, 30, 35, 40, 45, 50]
folder_names = [f"{truncation_value}_trunc" for truncation_value in truncation_values]

# In path_to_output, create folders from folder_names
for folder_name in folder_names:
    os.makedirs(path_to_output + folder_name, exist_ok=True)
```

```python
# Read every image from the input folder and call the executable with the image as input
for filename in os.listdir(path_to_input):
    if filename.endswith(".pgm"):
        
        input_image_path = path_to_input + filename
        
        output_image_path = ""
        for truncation_value in truncation_values:
            output_image_path = path_to_output + f"{truncation_value}_trunc/" + filename
            cmd = f"./susan{truncation_value} {input_image_path} {output_image_path} -e"
            os.system(cmd)        
        
        
        print(f"Processed {input_image_path}...")
```
